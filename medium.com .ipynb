# Descriptive Statistics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = [2, 4, 6, 6, 7, 8, 10, 12, 14, 16, 18, 20]

# Measures of central tendency
mean = np.mean(data)
median = np.median(data)
mode = pd.Series(data).mode()[0]  # mode() returns a Series, so we extract the first value

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

# Measures of dispersion
variance = np.var(data)
std_deviation = np.std(data)

print("Variance:", variance)
print("Standard Deviation:", std_deviation)

# Percentiles and quartiles
percentile_25 = np.percentile(data, 25)
percentile_50 = np.percentile(data, 50)
percentile_75 = np.percentile(data, 75)

print("25th Percentile (Q1):", percentile_25)
print("50th Percentile (Median/Q2):", percentile_50)
print("75th Percentile (Q3):", percentile_75)

# Skewness and kurtosis
skewness = pd.Series(data).skew()
kurtosis = pd.Series(data).kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

# Histogram and frequency distribution
plt.hist(data, bins='auto', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# Probability:

```python
import numpy as np
from scipy.stats import binom, norm, poisson

# Basic Concepts of Probability
sample_space = ['Heads', 'Tails']
events = {'A': ['Heads'], 'B': ['Tails']}
outcomes = len(sample_space)

print("Sample Space:", sample_space)
print("Events:", events)
print("Number of Outcomes:", outcomes)

# Probability Rules
prob_A = 0.4
prob_B = 0.6
prob_A_and_B = 0.2

# Addition Rule
prob_A_or_B = prob_A + prob_B - prob_A_and_B

# Multiplication Rule
prob_A_and_B_mult = prob_A * prob_B

print("Probability of A or B:", prob_A_or_B)
print("Probability of A and B (Multiplication Rule):", prob_A_and_B_mult)

# Conditional Probability
prob_A_given_B = prob_A_and_B / prob_B

print("Conditional Probability of A given B:", prob_A_given_B)

# Bayes' Theorem
prior_A = 0.3
likelihood_B_given_A = 0.7
evidence_B = 0.5

posterior_A_given_B = (likelihood_B_given_A * prior_A) / evidence_B

print("Posterior Probability of A given B (Bayes' Theorem):", posterior_A_given_B)

# Probability Distributions
n = 10
p = 0.3
x = np.arange(0, n+1)

# Binomial Distribution
binomial_dist = binom.pmf(x, n, p)

# Normal Distribution
mu = 5
sigma = 2
normal_dist = norm.pdf(x, mu, sigma)

# Poisson Distribution
lambda_ = 3
poisson_dist = poisson.pmf(x, lambda_)

print("Binomial Distribution:", binomial_dist)
print("Normal Distribution:", normal_dist)
print("Poisson Distribution:", poisson_dist)
```

# Inferential Statistics:

```python
import numpy as np
from scipy.stats import ttest_rel, ttest_ind, f_oneway, chi2_contingency

# Hypothesis Testing
sample1 = [1, 2, 3, 4, 5]
sample2 = [2, 4, 6, 8, 10]

# Null Hypothesis: The means of the two samples are equal
# Alternative Hypothesis: The means of the two samples are not equal

# Paired t-test (related samples)
t_statistic, p_value = ttest_rel(sample1, sample2)

print("Paired t-test:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Unpaired t-test (independent samples)
t_statistic, p_value = ttest_ind(sample1, sample2)

print("Unpaired t-test:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Analysis of Variance (ANOVA)
sample1 = [1, 2, 3, 4, 5]
sample2 = [2, 4, 6, 8, 10]
sample3 = [3, 6, 9, 12, 15]

# Null Hypothesis: The means of all groups are equal
# Alternative Hypothesis: At least one group mean is different

f_statistic, p_value = f_oneway(sample1, sample2, sample3)

print("ANOVA:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Chi-square Test
observed = np.array([[10, 20, 30], [15, 25, 35]])
expected = np.array([[20, 20, 20], [20, 20, 20]])

# Null Hypothesis: There is no association between the variables
# Alternative Hypothesis: There is an association between the variables

chi2_statistic, p_value, _, _ = chi2_contingency(observed)

print("Chi-square Test:")
print("Chi2-statistic:", chi2_statistic)
print("p-value:", p_value)
```

# Regression Analysis:

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Simple Linear Regression
data = pd.read_csv('data.csv')
X = data['independent_variable']
Y = data['dependent_variable']

X = sm.add_constant(X)  # Adding a constant intercept term

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)

print(model.summary())

# Multiple Linear Regression
X = data[['independent_variable1', 'independent_variable2']]
Y = data['dependent_variable']

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)

print(model.summary())

# Logistic Regression (Binary)
X = data[['independent_variable1', 'independent_variable2']]
Y = data['binary_target']

model = LogisticRegression()
model.fit(X, Y)
predictions = model.predict(X)

print(classification_report(Y, predictions))

# Logistic Regression (Multinomial)
X = data[['independent_variable1', 'independent_variable2']]
Y = data['multinomial_target']

model = LogisticRegression(multi_class='multinomial')
model.fit(X, Y)
predictions = model.predict(X)

print(classification_report(Y, predictions))
```

# Time Series Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Sample data
dates = pd.date_range(start='2023-01-01', periods=100, freq='M')
values = np.random.randint(50, 200, size=100)

# Creating a time series dataframe
data = pd.DataFrame({'Date': dates, 'Value': values})
data.set_index('Date', inplace=True)

# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Sample Time Series Data')
plt.show()

# Smoothing techniques - Moving Averages
moving_avg = data['Value'].rolling(window=5).mean()

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Value'], label='Original Data')
plt.plot(data.index, moving_avg, label='Moving Average')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Moving Average Smoothing')
plt.legend()
plt.show()

# Smoothing techniques - Exponential Smoothing
model = ExponentialSmoothing(data['Value'], trend='additive', seasonal=None)
fitted_model = model.fit()

forecast = fitted_model.forecast(steps=12)

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Value'], label='Original Data')
plt.plot(forecast.index, forecast, label='Exponential Smoothing')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Exponential Smoothing Forecast')
plt.legend()
plt.show()

# ARIMA model
model = ARIMA(data['Value'], order=(1, 1, 1))
fitted_model = model.fit()

forecast = fitted_model.forecast(steps=12)

plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Value'], label='Original Data')
plt.plot(forecast.index, forecast, label='ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```

# Linear Algebra:

```python
import numpy as np

# Matrix operations
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Matrix addition
matrix_sum = matrix1 + matrix2

print("Matrix Addition:")
print(matrix_sum)
print()

# Matrix subtraction
matrix_diff = matrix1 - matrix2

print("Matrix Subtraction:")
print(matrix_diff)
print()

# Matrix multiplication
matrix_product = np.matmul(matrix1, matrix2)

print("Matrix Multiplication:")
print(matrix_product)
print()

# Systems of linear equations
# 2x + 3y = 10
# 4x + 5y = 20
coefficients = np.array([[2, 3], [4, 5]])
constants = np.array([10, 20])

solutions = np.linalg.solve(coefficients, constants)

print("Solutions to the System of Linear Equations:")
print("x =", solutions[0])
print("y =", solutions[1])
print()

# Determinants
matrix = np.array([[1, 2], [3, 4]])

determinant = np.linalg.det(matrix)

print("Determinant of the Matrix:")
print(determinant)
print()

# Eigenvalues and Eigenvectors
matrix = np.array([[1, 2], [2, 4]])

eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("Eigenvalues:")
print(eigenvalues)
print()

print("Eigenvectors:")
print(eigenvectors)
```

# Optimization

```python
from scipy.optimize import linprog

# Linear Programming
# Maximize: 3x + 4y
# Subject to: x + y <= 5, x >= 0, y >= 0
c = [-3, -4]
A = [[1, 1]]
b = [5]
x_bounds = (0, None)
y_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds])

print("Linear Programming:")
print("Optimal Solution:", res.x)
print("Optimal Objective Value:", res.fun)
print()

from scipy.optimize import minimize

# Non-linear Programming
# Minimize: (x-2)^2 + (y-3)^2
# Subject to: x + y >= 3
# Initial guess: x=0, y=0
def objective(x):
    return (x[0] - 2) ** 2 + (x[1] - 3) ** 2

def constraint(x):
    return x[0] + x[1] - 3

initial_guess = [0, 0]
bounds = [(None, None), (None, None)]
constraint = {'type': 'ineq', 'fun': constraint}

res = minimize(objective, initial_guess, bounds=bounds, constraints=constraint)

print("Non-linear Programming:")
print("Optimal Solution:", res.x)
print("Optimal Objective Value:", res.fun)
print()

from scipy.optimize import minimize

# Constraint Optimization
# Minimize: x^2 + y^2
# Subject to: x + y = 1
# Initial guess: x=0, y=0
def objective(x):
    return x[0] ** 2 + x[1] ** 2

def constraint(x):
    return x[0] + x[1] - 1

initial_guess = [0, 0]
constraint = {'type': 'eq', 'fun': constraint}

res = minimize(objective, initial_guess, constraints=constraint)

print("Constraint Optimization:")
print("Optimal Solution:", res.x)
print("Optimal Objective Value:", res.fun)
```

# Discrete Mathematics:

```python
import itertools
import networkx as nx
import matplotlib.pyplot as plt

# Set Theory
set1 = {1, 2, 3}
set2 = {2, 3, 4}

# Union of sets
union_set = set1.union(set2)

print("Union of Sets:")
print(union_set)
print()

# Intersection of sets
intersection_set = set1.intersection(set2)

print("Intersection of Sets:")
print(intersection_set)
print()

# Combinatorics (Permutations and Combinations)
elements = [1, 2, 3]
permutations = list(itertools.permutations(elements))
combinations = list(itertools.combinations(elements, 2))

print("Permutations:")
for perm in permutations:
    print(perm)
print()

print("Combinations:")
for comb in combinations:
    print(comb)
print()

# Graph Theory
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])

# Graph visualization
nx.draw(G, with_labels=True)
plt.title("Graph")
plt.show()

# Connectivity
is_connected = nx.is_connected(G)

print("Graph Connectivity:")
print("Is the graph connected?", is_connected)
print()
```

# Data Manipulation and Analysis:

```python
import numpy as np
import pandas as pd

# Data Cleaning and Preprocessing
data = pd.DataFrame({'A': [1, 2, np.nan, 4],
                     'B': [5, 6, 7, 8],
                     'C': [9, np.nan, 11, 12]})

# Handling missing values
data_cleaned = data.dropna()

print("Data Cleaning:")
print("Original Data:")
print(data)
print()
print("Cleaned Data:")
print(data_cleaned)
print()

# Exploratory Data Analysis (EDA)
# Descriptive statistics
data_summary = data.describe()

print("Exploratory Data Analysis:")
print("Summary Statistics:")
print(data_summary)
print()

# Data Transformation (Normalization, Standardization)
# Normalization
data_normalized = (data - data.min()) / (data.max() - data.min())

print("Data Transformation:")
print("Normalized Data:")
print(data_normalized)
print()

# Database Querying (SQL)
# Querying a Pandas DataFrame using SQL-like syntax
query = "SELECT * FROM data WHERE A > 2"

data_query = pd.read_sql_query(query, con=data)

print("Database Querying:")
print("Query Result:")
print(data_query)
print()

# Data Wrangling using Pandas
# Grouping and calculating mean
data_grouped = data.groupby('B').mean()

print("Data Wrangling:")
print("Grouped Data:")
print(data_grouped)
```
